**Task 2**

```python
# Overall survival rate (percentage of passengers who survived)
survival_rate = df['Survived'].mean() * 100
print(f"Survival Rate: {survival_rate:.2f}%")
```

```python
# Distribution of Survived Variable (Count Plot Visualization)
sns.countplot(x='Survived', data=df)
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()
```

```python
# Passenger Class (Pclass) Distribution
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
# Identify Class with Most Passengers
df['Pclass'].value_counts()
```

**Task 3**

Survival by Sex

```python
# Crosstabulation
pd.crosstab(df['Sex'], df['Survived'])
```

```python
# Visualization
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Sex")
plt.show()
```

Survival by Passenger Class (Pclass)

```python
# Survival Rate Calculation
df.groupby('Pclass')['Survived'].mean()
```

```python
# Visualization
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

Survival by Age

```python
# Distribution Plot
sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True)
plt.title("Age Distribution by Survival")
plt.show()
```

Survival by Port of Embarkation (Embarked)

```python
# Survival rate
df.groupby('Embarked')['Survived'].mean()
```

```python
# Visualization
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title("Survival Rate by Embarkation Port")
plt.show()
```

**Task 4**

The exploratory data analysis indicates that survival on the Titanic was strongly influenced by gender, passenger class, and age. Females had a significantly higher survival rate than males, suggesting priority during evacuation. Passenger class also played a critical role, with 1st class passengers having much higher survival probabilities compared to those in 3rd class. Additionally, younger passengers—particularly children—had better survival chances than older individuals. Overall, Sex, Pclass, and Age emerge as the strongest predictors of survival in this dataset.
